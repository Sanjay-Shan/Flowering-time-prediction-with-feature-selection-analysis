```python
# Hereby we have defined the Independent variables and the response variable
X=data.loc[:, data.columns !='DtoA']
y=data['DtoA']
print(X.shape,y.shape)
```

```python
# Now that we completely understand the data well, lets enter the phase of data modelling
# Since this data has a total of 7393 variables associated with the target, we will first
# try to run the variable selection followed by data modelling
# 1. Pearson's Correlation Coefficient
# 2. Spearman's Correlation

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression # Is a correlation calculator
from scipy.stats import spearmanr
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error,r2_score

#first let's get the required number of varibles for modelling through pearson's correlation
k=[20,40,200,800,1000,2000,4000,5000]
lmse=[]
lr2=[]
for i in k:
    corr_data = SelectKBest(f_regression, k=i).fit(X, y)
    indices=corr_data.get_support()
    corr_features= X.columns.values[indices]

    corr_features_data= data.loc[:,corr_features]
    corr_feature_label=data['DtoA']

    # stdX = StandardScaler().fit_transform(corr_features_data)
    X_train, X_test, y_train, y_test = train_test_split(corr_features_data,corr_feature_label, test_size=0.2, random_state=1)

    corr_regr = RandomForestRegressor()

    corr_reg = corr_regr.fit(X_train,y_train)
    corr_pred = corr_regr.predict(X_test)

    #Get the required metric to understand performance of the trained model
    R2 = r2_score(y_test, corr_pred)
    mse = mean_squared_error(y_test, corr_pred)
    lmse.append(mse)
    lr2.append(R2)

    table=pd.DataFrame({'cols':['R2','MSE'],'val':[R2,mse]})
    print(table)

print(lmse)
print(lr2)

```

```python
# Lets perform the same with spearman's correlation 

k=[20,40,200,800,1000,2000,4000,5000]
Pmse=[]
Pr2=[]

spear_data=[]
for i in X.columns:
    spear,_= spearmanr(data['DtoA'],X[i])
    spear_data.append([i,round(spear,2)])

for i in k:
    spear=sorted(spear_data,key=lambda x:x[1],reverse=True)[:i]

    # get the names of the features
    spear_features=[]
    for a,b in spear:
        spear_features.append(a)

    spear_features_data=data.loc[:,spear_features]
    spear_feature_label=data['DtoA']

    # stdX = StandardScaler().fit_transform(spear_features_data)
    X_train, X_test, y_train, y_test = train_test_split(spear_features_data,spear_feature_label, test_size=0.2, random_state=1)

    spear_regr = RandomForestRegressor()

    spear_reg = spear_regr.fit(X_train,y_train)
    spear_pred = spear_regr.predict(X_test)

    #Get the required metric to understand performance of the trained model
    R2 = r2_score(y_test, spear_pred)
    mse = mean_squared_error(y_test, spear_pred)
    Pmse.append(mse)
    Pr2.append(R2)

    table=pd.DataFrame({'cols':['R2','MSE'],'val':[R2,mse]})
    print(table)

print(Pmse)
print(Pr2)
```
